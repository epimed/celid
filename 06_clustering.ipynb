# Clustering

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
```

```python
data_dir = 'data/'
```

```python
data = pd.read_csv(data_dir + 'TSG3_train_2386_samples_882_genes.csv', sep=';', index_col=0)
```

```python
y = data['target']
print(y.unique())
```

```python
X = data.drop(columns=['target'])
```

```python
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X, y)
X_scaled = pd.DataFrame(X_scaled, index=X.index, columns=X.columns)
```

```python
lut = {'bronchus_lung': 'royalblue', 'liver': 'yellow', 'heart': 'silver', 'brain': 'darkkhaki', 'esophagus': 'silver', 'connective_tissues': 'pink',
 'thyroid_gland': 'silver', 'prostate': 'silver', 'stomach': 'silver', 'kidney': 'silver', 'artery': 'silver', 'muscles': 'silver', 'blood': 'darkturquoise',
 'fibroblast': 'chartreuse', 'skin': 'silver', 'breast': 'crimson', 'central_nervous_system': 'lightsteelblue', 'uterus': 'silver', 'colon': 'silver',
 'testis': 'green'}
target_colors = y.map(lut)
g = sns.clustermap(X_scaled.T, col_colors=target_colors, row_cluster=False, method='ward')
```
